Matplotlib绘制三种常见图形

1. 折线图（Line Chart）

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
x = np.arange(1, 51)
y = np.random.rand(50)

plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

2. 条线图（Bar Chart）

```python
import matplotlib.pyplot as plt

# 示例数据
x = np.arange(1, 51)
y = np.random.rand(50)

plt.bar(x,y)
plt.title('Bar Chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

3. 散点图(Scatter Plot)

```python
import matplotlib.pyplot as py
import numpy as np

# 示例数据
x = np.arange(1, 51)
y = np.random.rand(50)

plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
